- The **ax.contour3D()** function creates three dimensional contour plot.

- It requires all the input data to be in the form of two-dimensional regular grids with the Z data evaluated at each point.

- Here we will show a three dimensional contour diagram of a three dimensional sinusoidal function.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def f(x,y):
  return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x,y)
Z = f(X,y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```
