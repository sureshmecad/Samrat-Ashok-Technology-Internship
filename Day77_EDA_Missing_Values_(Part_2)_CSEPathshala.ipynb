```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/CAR_Automobile_data.zip")
df.head()
```

```python
df['normalized-losses'].value_counts()
```

```python
df['horsepower'].value_counts()
```

```python
df.isna().sum()
```

```python
# Step1: Replace the missing values with nan
df['normalized-losses'].replace("?", np.nan, inplace=True)
df['horsepower'].replace("?", np.nan, inplace=True)

# Step2: change the datatype
df['normalized-losses'] = df['normalized-losses'].astype("float")
df['horsepower'] = df['horsepower'].astype("float")

# Step3: Get the mean
nlmean = df['normalized-losses'].mean()
hpmean = df['horsepower'].mean()

# Step4: Use the fillna method
df['normalized-losses'].fillna(nlmean, inplace=True)
df['horsepower'].fillna(hpmean, inplace=True)

df
```
