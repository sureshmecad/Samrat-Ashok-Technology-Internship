#### **What is a package?**

- A package contains all the files you need for a module.

- Modules are python code libraries you can include in your project.

#### **Check if PIP installed**

**1) List pip installed packages (with the default column formatting)**

```python
pip list
```

**2) List Installed Packages With Conda**

conda list

**3) List outdated packages with column formatting**

```python
pip list --outdated --format columns
```

**4) List packages that are not dependencies of other packages. Can be combined with other options**

```python
pip list --outdated --not-required
```
