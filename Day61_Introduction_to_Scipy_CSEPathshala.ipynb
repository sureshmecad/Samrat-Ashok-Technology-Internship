- **SciPy** in Python is an open-source library used for solving **mathematical, scientific, engineering, and technical problems.**

- It allows users to **manipulate the data and visualize the data** using a wide range of high-level Python commands. SciPy is built on the Python NumPy extention.

#### **SciPy Sub-packages**

 - **scipy.constants:** Physical and mathematical constants

 - **scipy.fftpack:** Fourier transform

 - **scipy.integrate:** Integration routines

 - **scipy.interpolate:** Interpolation

 - **scipy.io:** Data input and output

 - **scipy.linalg:** Linear algebra routines

 - **scipy.optimize:** Optimization

 - **scipy.signal:** Signal processing

 - **scipy.sparse:** Sparse matrices

 - **scipy.spatial:** Spatial data structures and algorithms

 - **scipy.special:** Any special mathematical functions

 - **scipy.stats:** Statistics
