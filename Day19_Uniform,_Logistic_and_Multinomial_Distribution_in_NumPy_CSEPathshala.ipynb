#### **1) Uniform Distribution**

- Used to describe probability where every event has equal chances of occuring.

- Ex: Generation of random numbers. It has three parameters:

 - a - lower bound --> default 0.0

 - b - upper bound --> default 1.0

 - size - The shape of the returned array

```python
import numpy as np
from numpy import random
```

```python
# create a 2x3 uniform distribution

x = random.uniform(size=(2,3))
print(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualization of uniform distribution

sns.distplot(random.uniform(size=1000), hist=False)
plt.show()
```

#### **2) Logistic Distribution**

- Logistic Distribution is used to describe growth. Used extensively in machine learning in logistic regression, neural networks etc.

- It has three parameters:

 - loc - mean, where the peak is Default 0.

 - scale - Standard Deviation, the flatness of distribution Default 1

 - size - The shape of the returned array

```python
# Draw 2x3 samples from a logistic distribution with mean at 1 and standard deviation 2

x = random.logistic(loc=1, scale=2, size=(2,3))
print(x)
```

```python
x = random.logistic(size=(2,3))
print(x)
```

```python
# Visualisation of Logistic Regression

sns.distplot(random.logistic(size=1000), hist=False)
plt.show()
```

```python
sns.distplot(random.logistic(loc=2, scale=3, size=1000), hist=False)
plt.show()
```

#### **3) Difference b/n Logistic and Normal Distribution**

- **Both** distributions are **near identical**, but **logistic distribution** has **more area under the tails.**

- That means: it represents more possibility of occurrence of an event further away from mean.

- For **higher** value of **scale (standard deviation)** the **normal and logistic distributions** are **near identical** apart from the **peak**.

```python
sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')
plt.show()
```

#### **4) Multinomial Distribution**

- Multinomial distribution is a generalization of binomial distribution. It describes outcome of multinomial scenarios unlike binomial where scenarios must be only one of two. 

- For example: Blood type of a population, dice roll outcome.

- It has three parameters:

 - n - number of possible outcomes (e.g. 6 for dice roll)

 - pvals - list of probabilies of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]for dice roll) 

```python
# Draw out a sample for dice roll

x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
print(x)
```

**Note:**
1. Multinomial samples will NOT produce a single value! They will produce one value for each pval.

2. As they are **generalization of binomial distribution** their visual representation and similarity of normal distribution is same as that of multiple binomial distributions.

```python
# Visualisation of Multinomial Distribution

sns.distplot(random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], size=10), hist=False, label='multinomial')
plt.show()
```
