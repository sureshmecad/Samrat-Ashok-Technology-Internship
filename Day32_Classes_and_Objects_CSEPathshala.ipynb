- A class is a userdefined blueprint or prototype from which objects are created. Classes provide a means of bundling data and functionality together.

- Creating a new class creats a new type of object, allowing new instances of that type to be made.

```python
class animal:
  def charact(self):
    print('4 legs', '2 eyes', 'mammal')

i = 3
print(type(i))

dog = animal()
print(type(dog))
```

```python
animal.charact(dog)
```

```python
dog.charact()
```

```python
cat = animal()
cat.charact()
```

#### **Create a Class**

- To create a class, use the keyword **class**:

```python
class MyClass:
  x = 5
```

#### **Create Object**

- Now we can use the class named MyClass to create objects:

```python
# Create an object named p1, and print the value of x
p1 = MyClass()
print(p1.x)
```

```python
class computer():
  def config(self):
    print("i5, 16gb, 1TB")

comp1 = computer()
comp2 = computer()

computer.config(comp1)
computer.config(comp2)

comp1.config()
comp2.config()
```
